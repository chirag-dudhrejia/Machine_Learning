```python
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from kmeans import Kmeans
```

```python
centroids = [(-5,-5), (7,7), (-1, 1), (10, -10)]
cluster_std = [1,1,1,1]
data, y = make_blobs(n_samples=100, cluster_std=cluster_std, centers=centroids, n_features=2, random_state=2)
```

```python
# plt.scatter(X[:,0], X[:,1])
plt.scatter(data[:, 0], data[:, 1])
plt.show()
```

```python
km = Kmeans(clusters=4, max_iter=100)
y_means = km.fit_predict(data)

# print(y_means)

km.ploting(data, y_means)
```
